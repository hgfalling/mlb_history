```python
gdfs = []
for filename in file_list:
    gl_headers = s2.split("\n")
    gldf = pd.read_csv(filename,names=gl_headers)
    gdfs.append(gldf)

gdf = pd.concat(gdfs,axis=0)
```

```python
gdf['YEAR'] = gdf.GAME_DT.apply(lambda x: int(str(x)[0:4]))
gdf['GAME_ID'] = gdf.HOME_TEAM_ID + gdf.GAME_DT.astype(str) + gdf.GAME_CT.astype(str)
for side in ["HOME","AWAY"]:
    gdf[f'{side}_TB_CT'] = gdf[f'{side}_HITS_CT'] + gdf[f'{side}_2B_CT'] + 2*gdf[f'{side}_3B_CT'] + 3*gdf[f'{side}_HR_CT']
    gdf[f'{side}_PA_CT'] = gdf[f'{side}_AB_CT'] + gdf[f'{side}_BB_CT'] + gdf[f'{side}_HP_CT'] + gdf[f'{side}_SH_CT'] + gdf[f'{side}_SF_CT']
```

```python
gdf.to_hdf(RETROSHEET_GL_DIR + "/all_game_logs.h5","game_logs_df")
```
